#**Limpieza de Datos**

```python
df_clean = df.copy()

#Eliminar columnas completamente vacías
df_clean.dropna(axis=1, how='all', inplace=True)

print(df_clean.columns)

df_clean
```

```python
# Limpiar nombres de columnas
# elimina espacios en blanco (o tabulaciones, saltos de línea) al inicio y al final del texto
df_clean.columns = df_clean.columns.str.strip()

# Verificar que se limpió correctamente
print(df_clean.columns.tolist())  # ejemplo : 'Samples  ' → 'Samples'
```

```python
df_clean.set_index(df_clean["Samples"])
df_clean
```

```python
# Vamos a crear una nueva columna "Age_months" que extraiga la edad a partir del nombre del grupo experimental

def extraer_edad(grupo):
    if grupo.startswith('P0'):
        return 0  # recién nacidos
    elif grupo.startswith('6m'):
        return 6
    elif grupo.startswith('12m'):
        return 12
    elif grupo.startswith('18m'):
        return 18
    else:
        return None  # por si acaso hay un grupo no reconocido

# Aplicar la función para crear la nueva columna
df_clean['Age_months'] = df_clean['Experimental_group'].apply(extraer_edad)

# Mostrar ejemplo
df_clean[['Samples', 'Experimental_group', 'Age_months']].head(10)
```

```python
# Función para extraer sexo
def extraer_sexo(grupo):
    if grupo.startswith('P0'):
        return 'Unknown'  # Se desconocido
    elif '_M_' in grupo:
        return 'M'
    elif '_F_' in grupo:
        return 'F'
    else:
        return 'Unknown' # Se desconocido

# Crear la columna
df_clean['Sex'] = df_clean['Experimental_group'].apply(extraer_sexo)

# Mostrar ejemplo
df_clean[['Samples', 'Experimental_group', 'Age_months', 'Sex']]
```

```python
# Función para extraer genotipo
def extraer_genotipo(grupo):
    return grupo.split('_')[-1]  # tomar el último campo después del '_'

# Crear la columna
df_clean['Genotype'] = df_clean['Experimental_group'].apply(extraer_genotipo)

# Mostrar ejemplo
df_clean[['Samples', 'Experimental_group', 'Age_months', 'Sex', 'Genotype']]
```

```python
# Función para extraer el número de individuo
def extraer_numero_individuo(sample):
    return int(sample.split('_')[-1])  # tomar último campo y convertir a int

# Crear la columna
df_clean['Individual_number'] = df_clean['Samples'].apply(extraer_numero_individuo)

# Mostrar ejemplo
df_clean[['Samples', 'Experimental_group', 'Age_months', 'Sex', 'Genotype', 'Individual_number']]
```

```python
df_clean
```

```python
# Lista de columnas que quieres al inicio
columnas_inicio = [
    'Samples',
    'Experimental_group',
    'Age_months',
    'Sex',
    'Genotype',
    'Individual_number'
]

# Lista de las otras columnas (todas las demás que no están en columnas_inicio)
columnas_otras = [col for col in df_clean.columns if col not in columnas_inicio]

# Reordenar el DataFrame
df_clean = df_clean[columnas_inicio + columnas_otras]

# Verificar las primeras columnas
print(df_clean.columns[:10].tolist())  # muestra los Nombres de las 10 primeras columnas

df_clean
```

```python
# Eliminar columna Experimental_group
df_clean = df_clean.drop(columns=['Experimental_group'])
df_clean
```

```python
# Exportar a CSV
df_clean.to_csv('signature_clean_v2.csv', index=False)
```

```python
# Librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy.stats import ttest_ind, f_oneway
```

```python
# Cargar el dataset limpio
df = pd.read_csv('/content/signature_clean_v2.csv')

# Mostrar estructura
print(f"Número de filas y columnas: {df.shape}")

df.head()
```

```python
# Resumen de variables categóricas
print("Distribución por Genotype:")
print(df['Genotype'].value_counts())

print("\nDistribución por Age_months:")
print(df['Age_months'].value_counts())

print("\nDistribución por Sex:")
print(df['Sex'].value_counts())

# Crear un gráfico combinado con subplot: distribución por Genotype/Age_months y por Sex
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico de conteo por Genotype y Edad
sns.countplot(data=df, x='Age_months', hue='Genotype', ax=axes[0])
axes[0].set_title('Distribución por Edad y Genotipo')
axes[0].set_xlabel('Edad (meses)')
axes[0].set_ylabel('Número de muestras')
axes[0].tick_params(axis='x', rotation=45)

# Gráfico de conteo por Sex y Genotype
sns.countplot(data=df, x='Sex', hue='Genotype', ax=axes[1])
axes[1].set_title('Distribución por Sexo y Genotipo')
axes[1].set_xlabel('Sexo')
axes[1].set_ylabel('Número de muestras')

plt.tight_layout()
plt.show()
```

```python
# Selección de variables metabolómicas (todas las columnas desde la sexta en adelante)
X = df.iloc[:, 6:].values

# Estandarización
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar PCA para 4 componentes
pca = PCA(n_components=4)
pca_result = pca.fit_transform(X_scaled)

# Crear DataFrame con los resultados de las 4 componentes
df_pca_4 = pd.DataFrame({
    'PC1': pca_result[:, 0],
    'PC2': pca_result[:, 1],
    'PC3': pca_result[:, 2],
    'PC4': pca_result[:, 3],
    'Genotype': df['Genotype'],
    'Sex': df['Sex'],
    'Age_months': df['Age_months']
})

# Mostrar la varianza explicada acumulada para las 4 componentes
explained_variance_4 = pca.explained_variance_ratio_

print(explained_variance_4)

# Graficar PC1 vs PC2 con Genotype y Sex
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca_4, x='PC1', y='PC2', hue='Genotype', style='Sex', s=100)
plt.title(f'PCA - Varianza explicada: PC1={explained_variance_4[0]:.2f}, PC2={explained_variance_4[1]:.2f}')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Obtener los nombres de las variables metabolómicas
feature_names = df.columns[6:]

# Crear un DataFrame con los loadings (coeficientes) para las primeras 4 componentes
loadings = pd.DataFrame(pca.components_.T,
                        index=feature_names,
                        columns=['PC1_loading', 'PC2_loading', 'PC3_loading', 'PC4_loading'])

# Calcular la magnitud total de contribución por variable a las 4 componentes
loadings['Magnitude'] = (loadings[['PC1_loading', 'PC2_loading', 'PC3_loading', 'PC4_loading']]**2).sum(axis=1)**0.5

# Ordenar por magnitud descendente para ver las más influyentes
top_contributors = loadings.sort_values(by='Magnitude', ascending=False).head(15)

# Mostrar los resultados
print("El Top de 15 variables con mayor contribución a las primeras 4 componentes del PCA:")
print(top_contributors)
```

```python
# Librerías necesarias
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy.stats import f_oneway

# Seleccionamos las variables metabolómicas (columnas 6 en adelante)
metabolitos = df.columns[6:]  # Ajusta si tus columnas de interés comienzan en otro lugar

# Lista para guardar resultados
resultados_anova = []

#Se crean los grupos según el Genotipo (C57, KI, 3xAD, etc.).
#Se eliminan posibles valores nulos (dropna()).
#Se aplica ANOVA de una vía para ver si hay diferencias estadísticas
# entre genotipos para ese metabolito.
for metabolito in metabolitos:
    # Preparar grupos para ANOVA
    grupos = [df[df['Genotype'] == g][metabolito].dropna() for g in df['Genotype'].unique()]

    # Calcular ANOVA
    anova_result = f_oneway(*grupos)

    # Guardar resultado ANOVA
    resultados_anova.append({
        'Metabolito': metabolito,
        'F_stat': anova_result.statistic,
        'p_value': anova_result.pvalue
    })

# Convertir a DataFrame
df_anova = pd.DataFrame(resultados_anova)

# Ordenar por p-valor
df_anova = df_anova.sort_values('p_value')

# Mostrar los top 15 metabolitos más significativos
df_anova.head(15)
```

Con base en los dos análisis de (PCA y ANOVA), las variables ideales para un análisis individual adicional son aquellas que:

* Tienen alta significancia estadística en el ANOVA (bajo valor p)
* Tienen alta contribución a los componentes principales del PCA (alto loading o magnitud)

#**Aprendizaje de Maquina**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# --- Preparar X (features) ---
X = df[metabolitos_significativos].values

# --- Preparar y (target) ---
# Etiqueta 1 para Alzheimer (3xAD), 0 para los otros
y = (df['Genotype'] == '3xAD').astype(int).values

# --- División en training / test ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# --- Entrenar modelo: RandomForestClassifier ---
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# --- Predicciones ---
y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:,1]
```

```python
# --- Accuracy ---
acc = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc:.2%}')

print("\n")
# --- Matriz de confusión ---
cm = confusion_matrix(y_test, y_pred)
print('Matriz de confusión:')
print(cm)

print("\n")
# --- Reporte de clasificación ---
print('Reporte de clasificación:')
print(classification_report(y_test, y_pred))

print("\n")
# --- ROC AUC ---
roc_auc = roc_auc_score(y_test, y_proba)
print(f'ROC AUC: {roc_auc:.2f}')

print("\n")
# --- Curva ROC ---
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Modelo Alzheimer')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score


# --- Diccionario para resultados ---
resultados = {}

# --- 1. Random Forest ---
clf_rf = RandomForestClassifier(n_estimators=100, random_state=42)
clf_rf.fit(X_train, y_train)
y_pred_rf = clf_rf.predict(X_test)
y_proba_rf = clf_rf.predict_proba(X_test)[:,1]

resultados['Random Forest'] = {
    'Accuracy': accuracy_score(y_test, y_pred_rf),
    'F1': f1_score(y_test, y_pred_rf),
    'ROC AUC': roc_auc_score(y_test, y_proba_rf)
}

# --- 2. Logistic Regression ---
clf_lr = LogisticRegression(max_iter=1000, random_state=42)
clf_lr.fit(X_train, y_train)
y_pred_lr = clf_lr.predict(X_test)
y_proba_lr = clf_lr.predict_proba(X_test)[:,1]

resultados['Logistic Regression'] = {
    'Accuracy': accuracy_score(y_test, y_pred_lr),
    'F1': f1_score(y_test, y_pred_lr),
    'ROC AUC': roc_auc_score(y_test, y_proba_lr)
}

# --- 3. SVM (con probabilidades) ---
clf_svm = SVC(kernel='rbf', probability=True, random_state=42)
clf_svm.fit(X_train, y_train)
y_pred_svm = clf_svm.predict(X_test)
y_proba_svm = clf_svm.predict_proba(X_test)[:,1]

resultados['SVM (RBF)'] = {
    'Accuracy': accuracy_score(y_test, y_pred_svm),
    'F1': f1_score(y_test, y_pred_svm),
    'ROC AUC': roc_auc_score(y_test, y_proba_svm)
}

# --- 4. Gradient Boosting ---
clf_gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
clf_gb.fit(X_train, y_train)
y_pred_gb = clf_gb.predict(X_test)
y_proba_gb = clf_gb.predict_proba(X_test)[:,1]

resultados['Gradient Boosting'] = {
    'Accuracy': accuracy_score(y_test, y_pred_gb),
    'F1': f1_score(y_test, y_pred_gb),
    'ROC AUC': roc_auc_score(y_test, y_proba_gb)
}
```

```python
# --- Convertir resultados a DataFrame para comparar ---
df_resultados = pd.DataFrame(resultados).T

# Mostrar ordenado por ROC AUC
df_resultados = df_resultados.sort_values(by='ROC AUC', ascending=False)

print('Comparación de modelos:')
display(df_resultados)
```
